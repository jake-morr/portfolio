```python
# Load common packages, load dataset, and preview the dataset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the main dataset
df = pd.read_csv('/kaggle/input/uber-ride-analytics-dashboard/ncr_ride_bookings.csv')

print("Shape (rows, cols):", df.shape)   # size of the dataset
print("\nFirst 5 rows:")
display(df.head())                       # preview rows
```

```python
# View info and data types

print("\nDataset Info:")
df.info()   
```

```python
# view summary statistics

print("\nSummary Stats (numerical):")
display(df.describe().T) 
```

```python
# check for missing values

print("\nMissing Values (count):")
print(df.isnull().sum())

print("\nMissing Values (%):")
print((df.isnull().mean() * 100).round(2))
```

```python
# view unique values

print("\nUnique Values per Column:")
print(df.nunique())
```

```python
# check for duplicates

print("\nDuplicate Rows:", df.duplicated().sum())
```
